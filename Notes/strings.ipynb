# Strings in C

The <string.h> header file in C provides functions to handle and manipulate strings (arrays of characters ending with a null character '\0').

Common String Functions

| Function             | Description                                                   |
| -------------------- | ------------------------------------------------------------- |
| `strlen(str)`        | Returns the length of a string                                |
| `strcpy(dest, src)`  | Copies one string into another                                |
| `strcat(dest, src)`  | Concatenates two strings                                      |
| `strcmp(str1, str2)` | Compares two strings                                          |
| `strrev(str)`        | Reverses a string (non-standard, available in some compilers) |


Explanation

* #include <string.h> → imports string functions.

* displayString(char str[]) → function receives a string (argument passing by reference).

* strlen(), strcpy(), strcat(), and strcmp() → demonstrate string operations.

* gets() → used for simple input (though unsafe; use fgets() in practice).

Applications of Strings

* Storing and manipulating text (e.g., names, sentences)

* Comparing user input with stored data

* Parsing and formatting data

* Implementing text-based menus and file handling

```python
%%writefile strings.c
#include <stdio.h>
#include <string.h>

// Function to display string details (argument passing example)
void displayString(char str[]) {
    printf("\nYou entered: %s", str);
    printf("\nLength of the string: %lu", strlen(str));

    // Copy string to another variable
    char copy[50];
    strcpy(copy, str);
    printf("\nCopied string: %s", copy);

    // Concatenate another string
    strcat(copy, " World!");
    printf("\nAfter concatenation: %s", copy);

    // Compare two strings
    if (strcmp(str, "Hello") == 0)
        printf("\nThe string is exactly 'Hello'.");
    else
        printf("\nThe string is not 'Hello'.");
}

int main() {
    char input[50];

    printf("Enter a string: ");
    /*
    if (fgets(input, sizeof(input), stdin) != NULL) {
        // Input was successfully read
    } else {
        // Error reading input or EOF reached
    }
    */
    gets(input);
    // Function call with string argument
    displayString(input);

    return 0;
}

```
